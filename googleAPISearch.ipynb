```python
import json

from googleapiclient.discovery import build
```

```python
pageLimit = 4
```

```python
service = build("customsearch", "v1",
            developerKey="your key")
```

```python
service
```

```python
startIndex = 1
response = []

for nPage in range(0, pageLimit):
    print ("Reading page number:",nPage+1)

    response.append(service.cse().list(
        q='names of search engines', #Search words
        cx='your key',  #CSE Key
        lr='lang_pt', #Search language
        start=startIndex
    ).execute())

    print(response)
    startIndex = response[nPage].get("queries").get("nextPage")[0].get("startIndex")
    
    with open('data.json', 'w') as outfile:
        json.dump(response, outfile)
```
